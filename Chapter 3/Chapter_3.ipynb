```python
from sklearn.datasets import fetch_openml
import numpy as np

# Mengunduh dataset MNIST
mnist = fetch_openml('mnist_784', version=1)
print(mnist.keys()) #

# Memisahkan fitur (X) dan label (y)
X, y = mnist["data"], mnist["target"]
y = y.astype(np.uint8) # Mengubah label menjadi integer

print("Bentuk X:", X.shape) # Output: (70000, 784)
print("Bentuk y:", y.shape) # Output: (70000,)
```

```python
# Memisahkan training dan test set
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:] #
```

```python
from sklearn.linear_model import SGDClassifier

# Membuat target untuk klasifikasi biner
y_train_5 = (y_train == 5) # True untuk semua angka 5, False untuk lainnya
y_test_5 = (y_test == 5) #

# Membuat dan melatih model SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_5) #
```

```python
from sklearn.model_selection import cross_val_score

# Mengevaluasi akurasi menggunakan cross-validation
print(cross_val_score(sgd_clf, X_train, y_train_5, cv=3, scoring="accuracy")) #

# Contoh classifier bodoh yang selalu memprediksi "bukan-5"
from sklearn.base import BaseEstimator
class Never5Classifier(BaseEstimator):
    def fit(self, X, y=None):
        return self
    def predict(self, X):
        return np.zeros((len(X), 1), dtype=bool)

never_5_clf = Never5Classifier()
print(cross_val_score(never_5_clf, X_train, y_train_5, cv=3, scoring="accuracy")) #
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

# Mendapatkan prediksi menggunakan cross-validation
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3) #

# Menghitung confusion matrix
print(confusion_matrix(y_train_5, y_train_pred)) #
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print("Precision:", precision_score(y_train_5, y_train_pred)) #
print("Recall:", recall_score(y_train_5, y_train_pred)) #

# F1 Score adalah rata-rata harmonik dari precision dan recall
print("F1 Score:", f1_score(y_train_5, y_train_pred)) #
```
