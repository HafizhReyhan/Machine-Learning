```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'netflix_titles.csv'
netflix_data = pd.read_csv(file_path)

# Set plot style
sns.set(style="whitegrid")

# 1. Distribution of Content Type (Movie vs TV Show)
plt.figure(figsize=(8, 5))
sns.countplot(data=netflix_data, x='type', palette='Set2')
plt.title('Distribution of Content Type (Movie vs TV Show)', fontsize=14)
plt.xlabel('Type of Content', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# 2. Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=netflix_data, x='rating', palette='Set3', order=netflix_data['rating'].value_counts().index)
plt.title('Distribution of Ratings', fontsize=14)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()

# 3. Distribution of Release Year
plt.figure(figsize=(10, 6))
sns.histplot(netflix_data['release_year'], kde=True, color='purple', bins=30)
plt.title('Distribution of Release Year', fontsize=14)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# 4. Distribution of Duration (Minutes or Seasons)
# Extract numeric value from 'duration' column
netflix_data['duration_numeric'] = netflix_data['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(10, 6))
sns.histplot(netflix_data['duration_numeric'], kde=True, color='green', bins=30)
plt.title('Distribution of Duration (in Minutes or Seasons)', fontsize=14)
plt.xlabel('Duration (Minutes or Seasons)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()

# 5. Correlation Matrix for Numerical Features
correlation_matrix = netflix_data[['release_year', 'duration_numeric']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix between Numerical Features', fontsize=14)
plt.show()

# 6. Duration vs Type of Content (Movie vs TV Show)
plt.figure(figsize=(10, 6))
sns.boxplot(data=netflix_data, x='type', y='duration_numeric', palette='Set2')
plt.title('Duration vs Type of Content (Movie vs TV Show)', fontsize=14)
plt.xlabel('Type of Content', fontsize=12)
plt.ylabel('Duration (Minutes or Seasons)', fontsize=12)
plt.show()

# 7. Duration vs Rating
plt.figure(figsize=(12, 6))
sns.boxplot(data=netflix_data, x='rating', y='duration_numeric', palette='Set3')
plt.title('Duration vs Rating', fontsize=14)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Duration (Minutes or Seasons)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()

# 8. Duration vs Country (Top 10 Countries)
top_countries = netflix_data['country'].value_counts().head(10).index
filtered_data = netflix_data[netflix_data['country'].isin(top_countries)]

plt.figure(figsize=(14, 8))
sns.boxplot(data=filtered_data, x='country', y='duration_numeric', palette='Set1')
plt.title('Duration vs Country (Top 10 Countries)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Duration (Minutes or Seasons)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.show()
```
